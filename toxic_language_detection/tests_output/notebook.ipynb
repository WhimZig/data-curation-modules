```python
from pprint import pprint

import pandas as pd

# Load the DataFrame
df = pd.read_csv("all_test_results.csv", delimiter=";")

df
```

```python
# Iterate over each row in the DataFrame
for _, row in df.iterrows():
    # Print the sentence
    print("Sentence:", row["sentence"])

    # Print the metadata
    print("Metadata:")
    pprint(row["metadata"])

    # New line for better readability between rows
    print("\n")
```
